### 6. Выдвинуть несколько гипотез по набору имеющихся переменных и проверить их с использованием t-test

H0 = Макс. скорость на бензите такая же, как на дизеле

```python
from scipy import stats

small_1 = df[df['fuel-type']=='gas']['highway-mpg']
small_2 = df[df['fuel-type']=='diesel']['highway-mpg']
print(stats.ttest_ind(small_1, small_2))
```

Вывод: отвергаем данную гипотезу, так как p-value равно 0.0059, что намного меньше стандартной альфа ошибки 0.05.
       Можем сказать, что имеется статистическое различие между выборками.

```python
df['make'].value_counts()
```

H0 = Сравним мощности у марок Mitsubishi и Audi 

```python
small_3 = df[df['make']=='mitsubishi']['horsepower']
small_4 = df[df['make']=='audi']['horsepower']
print(stats.ttest_ind (small_3, small_4))
```

Вывод: принимаем нулевую гипотезу, так как показатель p-value=0,19. Статистических различий между выборками не имеется.

```python
df['num-of-doors'].value_counts()
```

H0 = Цена машин с разным количеством дверей практически одинакова 

```python
small_5 = df[df['num-of-doors']=='four']['price']
small_6 = df[df['num-of-doors']=='two']['price']
print(stats.ttest_ind (small_5, small_6))
```

Вывод: принимаем нулевую гипотезу, как показатель p-value=0.79,что намного больше стандартной альфа ошибки 0.05. 
    Статистических различий между выборками не имеется.
